# Visualization using MatplotLib

```python
# Vertical and Horizontal Bar
import matplotlib.pyplot as plt
mylist = [1,2,3,4,5]
#plt.bar ( position, list, ....)
#plt.bar(mylist,mylist,align='center')
plt.barh(mylist,mylist,align='center')
plt.legend()
plt.show()
```

```python
# Stacked Bar
import matplotlib.pyplot as plt
mylist1 = [1,2,3,4,5]
mylist2 = [6,7,8,9,10]
mylist3 = [10,12,13,14,15]
#plt.bar ( position, list, ....)
plt.bar(mylist1,mylist3,label='one',color="red")
plt.bar(mylist1,mylist2,label='two',color="green")
plt.legend()
plt.show()
```

```python
# Multiple Bar charts
import matplotlib.pyplot as plt
mylist1 = [1,2,3,4,5]
mylist2 = [15,7,8,9,10]
mylist3 = [6,7,8,9,10]
mylist4 = [10,12,13,14,15]
#plt.bar ( position, list, ....)
plt.bar(mylist1,mylist2,label='one',color="red")
plt.bar(mylist3,mylist4,label='two',color="green")
plt.legend()
plt.xlabel("values")
plt.ylabel("heights")
plt.title("Two Bar Charts")
plt.show()
```

```python
import pandas as pd
data = pd.read_csv(filepath)
data.head()
```

```python
# Removed non-numeric columns
# Later on we will talk about One hot encoding or converign categorical into numerical

str_list = [] # empty list to contain columns with strings (words)
for colname, colvalue in data.iteritems():
    if type(colvalue[1]) == str:
         str_list.append(colname)
# Get to the numeric columns by inversion            
num_list = data.columns.difference(str_list) 
data_num = data[num_list]
#del movie # Get rid of movie df as we won't need it now
data_num.head()
```

Filling Null values and standardization

```python
"""
================
The Sankey class
================

Demonstrate the Sankey class by producing three basic diagrams.
"""
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.sankey import Sankey


# Example 1 -- Mostly defaults
# This demonstrates how to create a simple diagram by implicitly calling the
# Sankey.add() method and by appending finish() to the call to the class.
Sankey(flows=[1, -0.15, -0.60, -0.20, -0.15, -0.05, -0.50, -0.10],
       labels=['population', '', '', 'First', 'Second', 'Third', 'Fourth', 'Fifth'],
       orientations=[0, 1, 1, 1, 1, 1, 0, 1]).finish()
plt.title("The default settings produce a diagram like this.")
plt.show()
```
